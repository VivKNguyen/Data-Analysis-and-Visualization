### 3. Facial recognition with Eigenfaces

You can use Eigenfaces to perform facial recognition! Here is the idea:

1. Perform PCA on images of faces of people who you would like to be able to "learn" to recognize (creating a "database" of known faces). This is just like what you did in the main project.
2. Project a "query" image of a face that you would like to recognize into the PCA space established by the database of known faces (the number of PCA dimensions is up to you).  
3. Project each of the images in the face database into the same PCA space established by the database of known faces (the number of PCA dimensions is the same as before).
4. Treat the projected query image and each projected database images as vectors. Compute the Euclidean distance of the vector obtained by subtracting that of the query image and that of each database image.
5. Pick a match tolerance level. If any of the distances is less than your tolerance level, you have a match! If none of the distances is smaller than your tolerance, you don't have a match.

Extension:

I decided to try facial recognition using the pca methods developed in this project. I used the same face dataset from the main project.

Steps:
The first thing I did was create a new pca object called face_rec. I performed pca on that. I created a variable called eigenfaces which just includes the eigenvectors that came from that pca calculation. I projected back using 270 pcs because that's what I used for my 95% variance test, and I thought that would wield the most accurate results. I then reshaped just so I could have a visual of the 95% variance graphs. I chose a query image that would be used to test the recognition later.

Then I calculated weight vectors. These vectors represent the centered face pictures projected back onto the eigenfaces. It represents how close the face image is to the eigenface. 

Using those weight vectors, I found the distance using the euclidean method between the eigenface weights and the query weight and then found the index of that value. That index is best_match and provides the location of the face image that is closest to the query image. I then plotted the query and the best match image that the algorithm provided. It appears to be very successful and recognized the correct person.

```python
face_rec = pca_cov.PCA_COV(face_data)
face_rec.pca(vars)


images = face_rec.pca_then_project_back(270)
eigenfaces = face_rec.get_eigenvectors()[:270]
print(eigenfaces.shape)

face_plot(make_imgs(images), face_names)
```

```python
weights = eigenfaces @ (face_imgs_vec - np.mean(face_imgs_vec, axis=0)).T

query = face_imgs[38:39,:,]
N, M1, M2 = query.shape
query_vec = query.reshape(N, M1*M2)
query_weight = eigenfaces @ (query_vec - np.mean(face_imgs_vec,axis =0)).T
euclidean_distance = np.linalg.norm(weights - query_weight, axis=0)
best_match = np.argmin(euclidean_distance)

print(best_match)

fig, axes = plt.subplots(1,2,sharex=True,sharey=True,figsize=(8,6))

axes[0].imshow(query.reshape((64,64)), cmap="gray")
axes[0].set_title("Query")
axes[0].axis('off')
axes[1].imshow(face_imgs[best_match], cmap="gray")
axes[1].set_title("Best match")
axes[1].axis('off')
```
